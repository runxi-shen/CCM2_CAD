## Run CellBender

Parameters:
- `--cuda`: Use GPU acceleration
- `--learning-rate 0.00005`: Lower learning rate for stability
- `--checkpoint ckpt.tar.gz`: Save checkpoints for recovery if interrupted

```python
%%shell
eval "$(conda shell.bash hook)"
conda activate cellbender

# Process all samples
cellbender remove-background \
    --cuda \
    --input CCM2WTWeek0_filtered_feature_bc_matrix.h5 \
    --output CCM2WTWeek0_cellbender_output.h5 \
    --learning-rate 0.00005 \
    --checkpoint ckpt_WTWeek0.tar.gz

cellbender remove-background \
    --cuda \
    --input CCM2HetWeek0_filtered_feature_bc_matrix.h5 \
    --output CCM2HetWeek0_cellbender_output.h5 \
    --learning-rate 0.00005 \
    --checkpoint ckpt_HetWeek0.tar.gz

cellbender remove-background \
    --cuda \
    --input CCM2WTWeek12_filtered_feature_bc_matrix.h5 \
    --output CCM2WTWeek12_cellbender_output.h5 \
    --learning-rate 0.00005 \
    --checkpoint ckpt_WTWeek12.tar.gz

cellbender remove-background \
    --cuda \
    --input CCM2HetWeek12_filtered_feature_bc_matrix.h5 \
    --output CCM2HetWeek12_cellbender_output.h5 \
    --learning-rate 0.00005 \
    --checkpoint ckpt_HetWeek12.tar.gz
```
